```python
import PIL
import matplotlib.pyplot as plt
%matplotlib inline

class StatVisualizer:
    def __init__(self):
        with open("pose_random_index.txt", "r") as f:
            indices = f.readline()
            indices = indices.split(",")
            self.indices = [int(i) for i in indices]
            self.active = [self.map_us_to_miou(i+1) for i in range(len(indices))]
            
        miou_root = '../fashionAI/tools/segm_eval/results_11cls/'
        gfla_scores = np.load(miou_root+"gfla,square_scores.npy")
        ours_256 = np.load(miou_root+"our-large-square_scores.npy")
        adgan_scores = np.load(miou_root+"ADGAN_scores.npy")
        ours_176 = np.load(miou_root+"our-large_scores.npy")
        miou_fns = list(np.load(miou_root+"names.npy"))
        
        self.miou_scores = {}
        self.miou_scores['gfla'] = {miou_fns[i]: gfla_scores[i] for i in range(len(miou_fns))}
        self.miou_scores['ours256'] = {miou_fns[i]: ours_256[i] for i in range(len(miou_fns))}
        self.miou_scores['adgan'] = {miou_fns[i]: adgan_scores[i] for i in range(len(miou_fns))}
        self.miou_scores['ours176'] = {miou_fns[i]: ours_176[i] for i in range(len(miou_fns))}

    def map_us_to_miou(self, i):
        i = int(i) % 500
        return 'generated_%d.jpg'% (self.indices[i-1]+1)

    def get_images(self, i, roots=[], cat=True):
        fn = self.active[i]
        images = []
        #import pdb; pdb.set_trace()
        for root in roots:
            path = os.path.join(root, fn)
            image = PIL.Image.open(path)
            image = np.asarray(image)
            w = image.shape[1] 
            if w % 3 != 0:
                w = int(w/5)
                from_img, to_img, image = image[:,:w],image[:,w*2:w*3], image[:,w*4:]
            else:
                w = int(w/3)
                from_img, to_img, image = image[:,:w],image[:,w:w*2], image[:,w*2:]
            images.append(image)
        if not cat:
            return [from_img, to_img]+images
        image = np.concatenate([from_img, to_img]+images, 1)
        return image
    
    def set_human_scores(self, fn='stats.pkl'):
        with open(fn, 'rb') as f:
            stats = pickle.load(f)
            answers = stats['answer']
        questions_gfla = collections.defaultdict(list)
        questions_adgan = collections.defaultdict(list)
        for answer in answers:
            name, opt = answer.choice.split('; ')
            if not answer.task == 'posetrans':
                continue
            if 'gfla' in name:
                questions_gfla[self.map_us_to_miou(answer.question)].append(opt)
            else:
                questions_adgan[self.map_us_to_miou(answer.question)].append(opt)
        self.human_scores = {
            'gfla': questions_gfla,
            'adgan':  questions_adgan
        }
    
    def get_human_scores(self, cls, i):
        return self.human_scores[cls][self.active[i]]
    
    def get_miou_scores(self, cls, i):
        return self.miou_scores[cls][self.active[i]]
        
        
        
vs = StatVisualizer()
vs.set_human_scores()                                
 
```

```python
import json, os

HEAD = ['<!DOCTYPE html>','<html>', '<body>']
TAIL = ['</body>', '</html>']

class HTMLGenerator:
    def __init__(self):
        self.out_string = ""
        
    def _generate_head(self):
        out_string = '\n'.join(HEAD) + '\n'
        return out_string

    def _generate_tail(self):
        out_string = '\n'.join(TAIL) + '\n'
        return out_string
            

    def _generate_subject_section(self, exp_name='default'):
        out_string = '<h1> Visualization </h1>' + '\n'
        #out_string += '<h2>Exp Meta</h2>' + '\n'
        #out_string += '<p>Exp Name: %s</p>' % exp_name + '\n'
        #out_string += '<p>Epoch: %d</p>' % epoch + '\n'
        #out_string += '<p>Subset: %s </p>' % subset + '\n'
        return out_string


    
   
    def generate(self, out_path='tmp.html'):
        out_string = self._generate_head()
        #print(out_string)
        #subset_title = 'false_only_{} (if True, only those queries with R@10 = 0 displayed.)'
        out_string += self._generate_subject_section()
        out_string += self._generate_retrieval_table()
        out_string += self._generate_tail()
        with open(out_path, 'w') as f:
            f.write(out_string)

html = HTMLGenerator()
```
